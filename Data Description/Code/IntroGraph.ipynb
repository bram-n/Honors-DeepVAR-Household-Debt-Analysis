```python
countries = ['Canada', 'Australia', 'Korea, Rep. of', 'United States']
df2 = new_df[new_df['Country'].isin(countries)]

df2 = df2.rename(columns={'OBS_VALUE': 'Household Debt'})
```

```python
from datetime import datetime
def convert_quarter_to_start_datetime(quarter_str):
    year, quarter = quarter_str.split('-')
    year = int(year)
    quarter = int(quarter[1])  
    
    if quarter == 1:
        month = 1  # January
    elif quarter == 2:
        month = 4  # April
    elif quarter == 3:
        month = 7  # July
    elif quarter == 4:
        month = 10 # October

    return datetime(year, month, 1)

# Replace the 'TIME_PERIOD' column with the start dates
df2['TIME_PERIOD'] = df2['TIME_PERIOD'].apply(convert_quarter_to_start_datetime)
```

```python
df3 = df2[df2['TIME_PERIOD'] >= '2000-01-01']
df3
```

```python
line_styles = ['-', '--', '-.', ':']  

plt.figure(figsize=(10, 6))
for i, country in enumerate(df3['Country'].unique()):
    country_data = df3[df3['Country'] == country]
    plt.plot(country_data['TIME_PERIOD'], country_data['Household Debt'], 
             linestyle=line_styles[i % len(line_styles)], linewidth=2, label=country)
plt.axhline(y=85, color='r', linestyle='--', linewidth=1, label='Household Debt = 85')

plt.title('Household Debt/GDP Over Time')
plt.xlabel('Time Period')
plt.ylabel('Household Debt')
plt.xticks(rotation=45)

plt.legend()

plt.tight_layout()

plt.show()
```

```python
import pandas as pd
```

```python
def split_data(df):
    processed_dfs = []
    grouped = df.groupby('Country')
    
    for country, country_data in grouped:
        testdata = country_data.copy()
        
        testdata['Previous_GDP'] = testdata['GDP'].shift()
        testdata['Per_GDP'] = ((testdata['GDP'] - testdata['Previous_GDP']) / testdata['Previous_GDP']) * 100
        
        testdata['Previous_HHD'] = testdata['Household debt'].shift(1)
        testdata['Per_HHD'] = ((testdata['Household debt'] - testdata['Previous_HHD']) / testdata['Previous_HHD']) * 100
        
        processed_dfs.append(testdata)
    
    final_df = pd.concat(processed_dfs, ignore_index=True)
    
    below70 = final_df[final_df ["Household debt"] < 70].copy()
    above70 = final_df[final_df ["Household debt"] >= 70].copy()
    below70.to_excel("below70Data.xlsx", index=False)
    above70.to_excel("above70Data.xlsx", index=False)
    return below70
   
```
